```python
import open3d as o3d
import numpy as np 
import pandas as pd

df = pd.read_csv("codr_lidar_scan_data.csv")

# Conversion requires this to be in radians
# but servos work in degrees
altitude_rad = np.radians(df["altitude_angle_deg"])
azimuth_rad = np.radians(df["azimuth_angle_deg"])

# So that it's in m
# NOTE: Change the RPi driver so that it's converting to m there
# r = df["distance_cm"] / 100
r= df["distance_cm"]

# Spherical to Cartesian Conversion
# Using the picture version of Altitude and Azimuth
# There's other conversions depending on which angles are being
# examined
x = r * np.cos(altitude_rad) * np.cos(azimuth_rad)
y = r * np.cos(altitude_rad) * np.sin(azimuth_rad)
z = r * np.sin(altitude_rad)

# x, y, z are currently just numpy arrays
# We combine them into point cloud
point_cloud = np.column_stack((x,y,z))


# Use open3D to visualize
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(point_cloud)
o3d.visualization.draw_geometries([pcd])
```
